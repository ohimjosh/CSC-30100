```python
x = [1950,1951,1952,1953]
y = [34.72,34.66,34.79,34.85]
#x = np.array(list(sample_data.Date))
#y = np.array(list(sample_data.Price))

def interpolation(x, y, estimation, size):
    result = 0
    for i in range(size):
        tmp = 1.0
        for j in range(size):
            if i != j:
                tmp *= (estimation - x[j]) / (x[i] - x[j])
        result += tmp * y[i]
    return result


print("Lagrange Gold Price")
print("At year 1955, price:", interpolation(x,y,1955,4))
print("At year 1958, price:", interpolation(x,y,1958,4))
print("At year 1960, price:", interpolation(x,y,1960,4))
print("At year 1965, price:", interpolation(x,y,1965,4))

print("At year 2000, price:", interpolation(x,y,2000,4))
print("At year 2010, price:", interpolation(x,y,2010,4))
```

```python
#x = [1950,1951,1952,1953]
#y = [34.72,34.66,34.79,34.85]

#remember index start at 0
#n = 3                  #size of the list
#a= 4               #point we are estimating            

x = [1950,1951,1952,1953]
y = [34.72,34.66,34.79,34.85]

size = len(x)-1
def newton(x,y,size,est):
    c = [0]*(size+1)
    for j in range(0,size+1):
        c[j]=y[j]
    for k in range(1,size+1):
        for j in range(size,k-1,-1):
            c[j] = (c[j]-c[j-1])/(x[j]-x[j-k])

    p=c[size]
    for j in range(size-1,-1,-1):
        p=c[j]+(est-x[j])*p          
    return p

print("Newton Gold Price")
print("At year 1955, price:", newton(x,y,size,1955))
print("At year 1958, price:", newton(x,y,size,1958))
print("At year 1960, price:", newton(x,y,size,1960))
print("At year 1965, price:", newton(x,y,size,1965))

print("At year 2000, price:", newton(x,y,size,2000))
print("At year 2010, price:", newton(x,y,size,2010))
```
